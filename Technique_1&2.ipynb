# Text Mining

### Pre-processing reviews

```python
#Create function to remove punctations

text = 'This is a sampled sentence, showing off the process of the word filtration system.'

def punk_remove(text):
    words = nltk.word_tokenize(text)
    new_words = [word.lower() for word in words if word.isalnum()]
    return new_words

rempunk_text = punk_remove(text)
print(rempunk_text)
```

```python
#Create function to filter a review

stop_words = set(stopwords.words('english'))  
def filter_text(text):
    word_tokens = text
    filtered_text = [w for w in word_tokens if not w in stop_words]
    filtered_text = []  
    for w in word_tokens:  
        if w not in stop_words:  
            filtered_text.append(w)
    return filtered_text

filtered_text = filter_text(rempunk_text)
print(filtered_text) 
```

```python
#Create function to perform stemming a list of words

ps = PorterStemmer()
def stem_text(text_list):
    stemmed_words=[]
    for w in text_list:
        stemmed_words.append(ps.stem(w))
    return stemmed_words

stemmed_words = stem_text(filtered_text)
print(stemmed_words)
    
```

```python
#Create function to lemmatize a list of words

lem = WordNetLemmatizer()
def lemmatize_text(text_list):
    lemmatized_words=[]
    for w in text_list:
        lemmatized_words.append(lem.lemmatize(w,'v'))
    return lemmatized_words

lemmatized_words = lemmatize_text(stemmed_words)
print(lemmatized_words)
```

```python
#Process all reviews and fill 'review processed' column
#df_temp = df.head(1000)
df['review processed'] = np.nan
for i in range(df.shape[0]):
    df['review processed'][i] = lemmatize_text(filter_text(stem_text(punk_remove(df['review text'][i]))))
    
df
```

# Split Training and Testing Data

```python
import random

X = df_vect
# X.insert(0,'review rating',test_df['review rating'])
Y = test_df["ranking"]

print('Data size:',X.shape,Y.shape)

def train_test_split(X,Y,test_size=.20,rand=False):
    step=int(1/test_size)
    X_test=X.iloc[::step,:]
    Y_test=Y.iloc[::step]
    
    if rand:
        test_ind=random.sample(range(X.shape[0]), int(100*test_size))
        X_test=X.iloc[test_ind,:]
        Y_test=Y.iloc[test_ind]
        
    X_train=X.drop(X_test.index)
    Y_train=Y.drop(Y_test.index)
    
    return(X_train,X_test,Y_train,Y_test)
```

# PCA

```python
(X_train,X_test,Y_train,Y_test) = train_test_split(X,Y)

pca = decomposition.PCA(n_components=3)
pca.fit(X_train)
X_train = pca.transform(X_train)
X_test = pca.transform(X_test)

print('Training size:',X_train.shape,Y_train.shape)
print('Testing size:',X_test.shape,Y_test.shape)
```

# Linear Regression

```python
def LR_train_test_error(X,Y,PCA_dim=None,use_PCA=True):
    #80:20 -> training:testing
    (X_train,X_test,Y_train,Y_test) = train_test_split(X,Y)
    
    #PCA
    if use_PCA:
        pca = decomposition.PCA(n_components=PCA_dim)
        pca.fit(X_train)
        X_train = pca.transform(X_train)
        X_test = pca.transform(X_test)
    
    #Linear Regression
    book_model = LinearRegression()
    book_model.fit(X=X_train, y=Y_train)
    
    #Mean Absolute Error
    Y_pred = book_model.predict(X=X_test)
    mae = (abs(Y_test-Y_pred)).mean()
    return mae,Y_pred

#Use sklearn method to find accuracy


temp = X.drop(X.iloc[::5,:].index).shape
max_n = min(temp[0], temp[1]) - 1
# print(max_n)
```

```python
#LR without PCA

(LR_mae,LR_pred) = LR_train_test_error(X,Y,use_PCA=False)
print('Linear Regression Mean Absolute Error without PCA:',LR_mae)
```

```python
#LR with PCA

err=[]
for n in range(1,max_n):
    (mae,LR_Y_pred) = LR_train_test_error(X,Y,PCA_dim=n)
    err.append(mae)
    
best_PCA_n = err.index(min(err))+1
print("Best n for PCA:", best_PCA_n)
print("Best error for PCA:", min(err))
```

```python
fig = plt.figure()
plt.plot([1,100],[1,100],label='Ideal Prediction Line')

# plt.plot(Y_test,LR_pred,'yo',label="LR Prediction (No PCA)")

# x = np.array(Y_test)
# y = np.array(LR_pred)
# m, b = np.polyfit(x, y, 1)
# plt.plot(x, m*x + b,'_y')


(LR_mae_PCA,LR_pred_PCA)=LR_train_test_error(X,Y,PCA_dim=best_PCA_n)
plt.plot(Y_test,LR_pred_PCA,'ro',label="LR Prediction (PCA)")

x = np.array(Y_test)
y = np.array(LR_pred_PCA)
m, b = np.polyfit(x, y, 1)
# plt.plot(x, m*x + b,'_r')

fig.suptitle('Linear Regression: Predictions')
plt.xlabel('True Book Ranking')
plt.ylabel('Predicted Book Ranking')
# plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```

```python
#KNN with PCA

(X_train,X_test,Y_train,Y_test) = train_test_split(X,Y)

pca = decomposition.PCA(n_components=best_PCA_n-1)
pca.fit(X_train)
X_train = pca.transform(X_train)
X_test = pca.transform(X_test)

mae_with_PCA = []
for k in range(1,70):
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(X_train, Y_train)
    Y_pred = knn.predict(X_test)
    mae = (abs(Y_test-Y_pred)).mean()
    mae_with_PCA.append(mae)
        
best_k_PCA = mae_with_PCA.index(min(mae_with_PCA))+1
print("Best k for KNN (with PCA):", best_k_PCA)
print("Best error for KNN (with PCA):", min(err))

mae_with_PCA = np.array(mae_with_PCA)
```
